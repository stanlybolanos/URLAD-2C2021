## Conexion a sql

```python
conexion = pyodbc.connect('Driver={SQL Server};'
'Server=localhost;'
'Database=Admisiones_DWH;'
'Trusted_Connection=yes;')
```

```python
DF_DB = pd.read_sql('select * from fact.examen e inner join dimension.carrera c on e.sk_carrera=c.sk_carrera',conexion)
DF_DB.head(5)
```

```python
df_ConteoExamenes=DF_DB.groupby('NombreFacultad').count()['ID_Examen'].reset_index(name='ConteoExamenes')
```

```python
plt.pie(df_ConteoExamenes['ConteoExamenes'],labels=df_ConteoExamenes.NombreFacultad,autopct='%1.1f%%')
```
